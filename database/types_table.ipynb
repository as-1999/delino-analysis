```python
import pymysql
import pandas as pd
import os
import json
import mysql.connector
```

```python
con = pymysql.connect(
    host = 'host_name',
    user = 'user_name',
    passwd = 'password',
    db = 'db_name')
cursor = con.cursor()
```

```python
file = os.path.abspath("new.json")
json_data = open(file).read()
json_obj = json.loads(json_data)
```

```python
query = """
CREATE TABLE IF NOT EXISTS types (id NVARCHAR(64), 
FOREIGN KEY (id) REFERENCES main(id), 
east_asian INT(1), 
italian INT(1),
food_buffet INT(1),
international INT(1),
doner_kebab INT(1),
sandwich INT(1),
refectory INT(1),
tabbakhi INT(1),
veg_food INT(1),
iranian INT(1),
special_food INT(1),
sea_food INT(1),
european_food INT(1),
french INT(1),
fastfood INT(1),
food_court INT(1),
lebanese INT(1),
fried_chicken INT(1),
mexican INT(1),
indian INT(1),
pizza INT(1),
pizza_sanwich INT(1),
chinese INT(1),
japanese INT(1),
coffeeshop INT(1),
turkish_kebab INT(1),
kebab_house INT(1),
grill INT(1),
gilaki INT(1)
);
"""
cursor.execute(query)
```
